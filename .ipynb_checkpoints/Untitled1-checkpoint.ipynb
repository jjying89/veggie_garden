```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs


def init_browser():
    # @NOTE: Replace the path with your actual path to the chromedriver
    executable_path = {'executable_path': 'C:/chromedriver'}
    return Browser('chrome', **executable_path, headless=False)


def scrape():
    
    browser = init_browser()
    mars_data = {}

    ## NASA Mars News
    # visit url
    news_url = 'https://mars.nasa.gov/news/'
    browser.visit(news_url)

    # read url html
    news_html = browser.html
    news_soup = bs(news_html, 'html.parser')

    # get news title
    news_title = news_soup.find_all('div', class_='content_title')[1].text
    

    # get news snippet
    news_p = news_soup.find('div', class_='article_teaser_body').text
    

    ## JPL Mars Image
    # visit url
    image_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
    browser.visit(image_url)

    # navigate website
    browser.links.find_by_partial_text('FULL IMAGE').click()
    browser.links.find_by_partial_text('more info').click()

    # read url html
    image_html = browser.html
    image_soup = bs(image_html, 'html.parser')


    # get image src
    image_link = image_soup.find_all('img')[3]['src']

    # get complete image url
    featured_image_url = 'https://www.jpl.nasa.gov' + image_link

    
    ## Mars Facts Table
    # visit url
    facts_url = 'https://space-facts.com/mars/'
    browser.visit(facts_url)

    # read html
    table = pd.read_html(facts_url)
    
    # load data into pandas
    mars_df = table[0]
    
    # convert to html table string
    mars_html_table = mars_df.to_html()
    
    # remove \n
    mars_html_table.replace('\n', '')
    

    ## Mars Hemispheres
    # visit url
    mars_hemi_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
    browser.visit(mars_hemi_url)

      
    # read url html
    hemi_html = browser.html
    hemi_soup = bs(hemi_html, 'html.parser')

    # get search results
    hemispheres = hemi_soup.find_all('div', class_='item')
    hemispheres[0]

    # get title and image url of each mars' hemispheres
    hemisphere_image_urls = []

    main_url = 'https://astrogeology.usgs.gov'

    for hemisphere in hemispheres:
    
        # get title of the hemisphere
        title = hemisphere.find('h3').text
    
        # get hemisphere href
        hemi_url = hemisphere.find('a', class_='itemLink product-item')['href']
        browser.visit(main_url + hemi_url)
    
        # visit hemisphere url link
        hemi_img_html = browser.html
        hemi_img_soup = bs(hemi_img_html, 'html.parser')
    
        # get hemisphere image href
        img_link = hemi_img_soup.find('div', class_='downloads')
        img_url = img_link.find('li').a['href']
    
        # combine results into dict and list
        hemisphere_image_dict = {}
        hemisphere_image_dict['title'] = title
        hemisphere_image_dict['img_url'] = img_url
    
        hemisphere_image_urls.append(hemisphere_image_dict)

    
    # close the browser after scraping
    browser.quit()

    # return results
    mars_data = {
        news_title,
        news_p,
        featured_image_url,
        mars_html_table
    }
    
    return mars_data
```
